```python
import requests
import pandas as pd
```

```python

# Etapa 1: Preparar o Terreno (Inputs e Parâmetros)
# conectar a api e colocar parametros
url = 'https://api.open-meteo.com/v1/forecast'

longitude=float(input('adicione a longitude desejada: '))
latitude=float(input('adicione a latitude desejada: '))


parametros = {
    'latitude' : latitude,
    'longitude' : longitude,
    'daily' : ['temperature_2m_max','precipitation_sum'],
    'timezone' : 'auto',
    'past_days' : 7
}

# Etapa 2: Coleta de Dados e Verificação
#Pedimos temperatura máxima e soma de chuva diária para os últimos 7 dias
#para eu coletar algo, chamamos a biblioteca requests com o atributo get(onde, o que)
# usar o get
dados_diarios = requests.get(url, params=parametros)
# if do status_code
if dados_diarios.status_code == 200:
# salvar no json
  df_clima = dados_diarios.json()
  # criar o dataframe com o pandas
  df_dados_clima= pd.DataFrame(df_clima['daily'])
  # EXTRA - mostrar no padrao dos estados unidos
  # (temperatura * 1.8) +32
  df_dados_clima['temperature_f'] = (df_dados_clima['temperature_2m_max']*1.8)+32
  #se quiser display(dataframe)
  display(df_dados_clima)

else:
  print(f'EPI nao identificado {dados_diarios.status_code}')
# Etapa 3: Salvar os Dados (O Arquivo CSV)

# salva o arquivo

df_dados_clima.to_csv('relatorio_prof_luiz.csv', index=False)
```
